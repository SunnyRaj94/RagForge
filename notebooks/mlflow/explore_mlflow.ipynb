# MLflow Experiment tracking Playground
This notebook demonstrates how to connect to MLflow tracking services, start logging runs, serialize parameter metrics, and inspect local directory configurations.

```python
import sys
from pathlib import Path

# Locate and append project root
def get_project_root() -> Path:
    current = Path.cwd()
    for _ in range(5):
        if (current / "pyproject.toml").exists():
            return current
        current = current.parent
    return Path.cwd()

sys.path.append(str(get_project_root().resolve()))
```

```python
import mlflow
from ragforge.config import MLFLOW_TRACKING_URI

print(f"Resolved MLflow Tracking URI: {MLFLOW_TRACKING_URI}")
```

## 1. Setup MLflow Tracking Context

```python
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)
mlflow.set_experiment("notebook-experiment-exploration")

print("MLflow experiment initialized!")
```

## 2. Log Parameters, Metrics, and Tags

```python
with mlflow.start_run() as run:
    # Log parameters (configuration choices)
    mlflow.log_param("embedding_model", "nomic-embed-text")
    mlflow.log_param("chunk_size", 800)
    mlflow.log_param("chunk_overlap", 100)
    
    # Log metrics (real-time measurements)
    mlflow.log_metric("indexing_latency_sec", 1.45)
    mlflow.log_metric("total_documents_processed", 12)
    mlflow.log_metric("chunk_count", 56)

    # Log tags (lineage attributes)
    mlflow.set_tag("environment", "jupyter-notebook")
    mlflow.set_tag("author", "Sunny Raj")

    print(f"Logged active run success! Run ID: {run.info.run_id}")
```

## 3. Query Run Statistics Programmatically

```python
experiment = mlflow.get_experiment_by_name("notebook-experiment-exploration")
runs = mlflow.search_runs(experiment_ids=[experiment.experiment_id])

print("=== Recorded Notebook Runs ===")
for idx, row in runs.iterrows():
    print(f"Run ID: {row['run_id']} | Latency: {row['metrics.indexing_latency_sec']:.2f}s | Chunk Size: {row['params.chunk_size']}")
```
